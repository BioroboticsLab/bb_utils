# Code with two ways to generate placement markers to put on the hive

```python
import bb_utils.fiducial
import matplotlib.pyplot as plt
```

```python
# Parameters
w = 4  # Number of columns
h = 4  # Number of rows
figsize=(32,32)

# Generate the marker sheet
bb_utils.fiducial.plot_sheet(w=w, h=h, figsize=figsize)
plt.savefig('fiducial_placement_marker_grid_sheet.pdf')
```

```python
import itertools
fig, axes = plt.subplots(w, h, figsize=figsize)
axes = itertools.chain(*axes)

for i, ax in enumerate(axes):
    bb_utils.fiducial.plot_marker(ax, inverse=(i % 2 == 0))

plt.savefig('fiducial_placement_marker_grid.pdf')
```
